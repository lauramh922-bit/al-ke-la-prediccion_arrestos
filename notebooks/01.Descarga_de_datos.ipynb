# CARGA Y CONSOLIDACION DEL DATASET

```python
import pandas as pd
```

```python

#cargamos el dataset
df_2009 = pd.read_csv("data/2009_1perc.csv", encoding= "latin-1")
df_2010 = pd.read_csv("data/2010_1perc.csv", encoding= "latin-1")

#consolidamos ambos set
# Verificar que ambos DataFrames tienen la misma estructura de columnas
if list(df_2009.columns) == list(df_2010.columns):
    # Concatenar los DataFrames
    df_consolidado = pd.concat([df_2009, df_2010], ignore_index=True)
    print('Los archivos han sido consolidados.')
else:
    print('Los DataFrames tienen estructuras de columnas diferentes')
```

```python
df_consolidado.to_csv("data/consolidado.csv", index=False)
```

## EXPLORACION BASICA

Obtenemos información básica del dataset: número de filas/columnas, tipos de datos, estadísticas básicas y número de vacios por variable.

```python
print(f"El DataFrame consolidado tiene {df_consolidado.shape[0]} filas y {df_consolidado.shape[1]} columnas.")

object_count = (df_consolidado.dtypes == 'object').sum()
print(f"Numero de variables tipo object: {object_count}")
object_count = (df_consolidado.dtypes == 'str').sum()
print(f"Numero de variables tipo string: {object_count}")
# Contar cuántas variables son de tipo 'integer'
integer_count = (df_consolidado.dtypes == 'int').sum() + (df_consolidado.dtypes == 'int64').sum()
print(f"Numbero de variables tipo entero: {integer_count}")
# Contar cuántas variables son de tipo 'float'
float_count = (df_consolidado.dtypes == 'float').sum() + (df_consolidado.dtypes == 'float64').sum()
print(f"Numbero de variables tipo float: {float_count}")
missing_values = df_consolidado.isnull().sum()
print(f"Numbero valores perdidos: {missing_values}")
total_nulos = df_consolidado.isnull().sum().sum()
print(f"El dataset tiene un total de {total_nulos} valores nulos.")
print("-------------")
print("Estadisticos del dataset")
df_consolidado.describe().T
```
