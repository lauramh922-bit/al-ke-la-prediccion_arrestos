```python
import plotly.express as px
import pandas as pd
import nbformat
```

```python
df_copy_18 = pd.read_csv("data/df_copy_18.csv")
```

```python
# 1. Preparar los datos para el gráfico de barras (conteo de frecuencias)
# Esto crea una tabla con el conteo de cada categoría
df_counts = df_copy_18['arstmade'].value_counts().reset_index()
df_counts.columns = ['arstmade', 'conteo']

# 2. Crear el Gráfico de Barras Interactivo
fig_bar = px.bar(df_counts,
                 x='arstmade',
                 y='conteo',
                 title='Distribución de arrestos realizados (Interactivo)',
                 labels={'arstmade': 'Arresto Realizado', 'conteo': 'Cantidad'},
                 color='arstmade',
                 color_discrete_sequence=['#3498db', '#e74c3c'],
                 template='plotly_white')

fig_bar.show()

# 3. Crear el Gráfico de Pastel (Pie Chart) Interactivo
fig_pie = px.pie(df_counts,
                 values='conteo',
                 names='arstmade',
                 title='Proporción de arstmade',
                 color_discrete_sequence=px.colors.qualitative.Pastel)

fig_pie.update_traces(textinfo='percent+label', pull=[0.1, 0]) # Resalta una rebanada
fig_pie.show()
```

Ambos años tienen una similar cantidad de procedimientos y arrestos, hubo un incremento de 149 casos más en 2010 que en 2009 y 52 más arrestos en 2010 que en 2009. Revisaremos la variable con un grafico por mes.

```python
# 1. Preparar los datos (Agrupamos por mes, año y arstmade)
df_agrupado = df_copy_18.groupby(['month', 'year', 'arstmade']).size().reset_index(name='conteo')

# 2. Crear el gráfico de barras apiladas
# En Plotly, usamos 'facet_col' o simplemente combinamos las variables en 'color'
fig = px.bar(df_agrupado,
             x='month',
             y='conteo',
             color='year', # Diferenciamos por año con colores
             pattern_shape='arstmade', # Podemos usar patrones para diferenciar el arresto
             title='Conteo de arstmade por Mes y Año (Interactivo)',
             labels={'month': 'Mes', 'conteo': 'Cantidad de Casos', 'year': 'Año', 'arstmade': 'Arresto'},
             template='plotly_white',
             text_auto=True) # Esto pone los números automáticamente

# 3. Configurar para que sea apilado (por defecto lo es, pero aseguramos)
fig.update_layout(barmode='stack', xaxis={'dtick': 1})

fig.show()
```

En 2009, junio y diciembre fueron los meses con menor cantidad de arrestos.
En 2010, en diciembre fué el mes de menor cantidad de arrestos.

Se revisa la relación de la variable respuesta en comportamiento con la raza (race)

```python
# 1. Agrupamos los datos por raza y estado de arresto
df_race = df_copy_18.groupby(['race', 'arstmade']).size().reset_index(name='conteo')

# 2. Creamos el gráfico de barras agrupadas
fig_race = px.bar(df_race,
                  x='race',
                  y='conteo',
                  color='arstmade',
                  barmode='group', # Barras una al lado de la otra
                  title='Frecuencia de Arrestos por Raza (Interactivo)',
                  labels={'race': 'Raza', 'conteo': 'Conteo de Arrestos', 'arstmade': 'Arresto Realizado'},
                  color_discrete_sequence=px.colors.qualitative.Set2, # Usamos la misma paleta Set2
                  text_auto=True, # Anotaciones automáticas
                  template='plotly_white')

# 3. Ajustamos el diseño
fig_race.update_layout(
    xaxis_title="Raza",
    yaxis_title="Conteo de Arrestos",
    legend_title="Arresto Realizado",
    font=dict(size=12)
)

fig_race.show()
```

En los años analizados, la mayoria de arrestos corresponde a la raza black (B) y en segundo lugar la raza hispana (Q)

Determinar si el procedimiento policial concluirá en alguna acción violenta.
Los atributos que tienen el prefijo pf (['pf_hands'],['pf_wall'], ['pf_grnd'],['pf_drwep'], ['pf_ptwep'],['pf_baton'],['pf_hcuff'], ['pf_pepsp'] y
['pf_other']) indican si hubo fuerza fisica utilizada por el oficial al momento del procedimiento, con la marca 'Y'.
Se genera una nueva variable llamada 'violencia' la cual sea 1 si en cualquiera de las 9 variables pf hubo alguna 'Y', y 0 en otro caso y se indica el porcentaje de casos que terminaron con violencia.

```python
# Lista con los nombres de las columnas con prefijo 'pf'
pf_columns = ['pf_hands', 'pf_wall', 'pf_grnd', 'pf_drwep', 'pf_ptwep',
              'pf_baton', 'pf_hcuff', 'pf_pepsp', 'pf_other']

# Agregar la columna 'violencia' basada en si alguna de las columnas 'pf' tiene 'Y'
df_copy_18['violencia'] = df_copy_18[pf_columns].apply(lambda row: 1 if 'Y' in row.values else 0, axis=1)

# Calcular el porcentaje de casos con violencia
violencia_porcentaje = df_copy_18['violencia'].mean() * 100

# Imprimir el porcentaje de casos que terminaron con violencia
print(f"El porcentaje de casos que terminaron con violencia es: {violencia_porcentaje:.2f}%")
```

```python
#creamos un dataframe que no contenga las columnas con prefijo pf, puesto fueron reemplazadas
df_arrest= df_copy_18.copy()
df_arrest= df_arrest.drop(pf_columns, axis=1)

#recodificamos la variable arstmade asi que dejamos solo arstmade_cod
df_arrest= df_arrest.drop('arstmade', axis=1)
display(df_arrest.info())
```
